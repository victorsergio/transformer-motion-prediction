```python
import math
def average_displacement_error(pred,ground):

    e = []

    for p,g in zip(pred,ground):
        e.append( math.sqrt (((p[0] - g[0]) ** 2) + ((p[1] - g[1]) ** 2)))

    return np.sum(e)/len(e),e[len(e)-1]   #ADE,FDE
```

```python
from joblib import load
from joblib import dump
import numpy as np
import torch

_X = load('/content/drive/Othercomputers/My Laptop/github-repositories/transformer-multi/output/_X_item.joblib')
_Y = load('/content/drive/Othercomputers/My Laptop/github-repositories/transformer-multi/output/_Y_item.joblib')
_Y_hat = load('/content/drive/Othercomputers/My Laptop/github-repositories/transformer-multi/output/_Y_hat_item.joblib')
```

```python
_X = torch.cat(_X, axis=1).permute(1,0,2).numpy()
_Y = torch.cat(_Y, axis=1).permute(1,0,2).numpy()
_Y_hat = torch.cat(_Y_hat, axis=1).permute(1,0,2).numpy()
```

```python
ade_list = list()
fde_list = list()

for t,p in zip(_Y,_Y_hat):
     ade,fde = average_displacement_error(p[:,:2],t[:,0:2])
     ade_list.append(ade)
     fde_list.append(fde)

avg_ade = round(sum(ade_list)/len(ade_list),3)
avg_fde = round(sum(fde_list)/len(fde_list),3)


print("avg_ade: "+ str(avg_ade))
print("avg_fde: "+str(avg_fde))

```
